```python
dfx= pd.concat([df1,df2,df3,df4,df5])
```

```python
dfx["LojaID"]=dfx["LojaID"].astype("object")
```

```python
dfx["Vendas"].fillna(dfx["Vendas"].mean(), inplace=True)
```

```python
dfx["Vendas"].fillna(0, inplace=True)
```

```python
dfx.dropna(subset=["Vendas"], inplace=True)
```

```python
dfx.dropna(how="all", inplace=True)
```

```python
dfx["Receitas"] = dfx["Vendas"]*(dfx["Qtde"])
```
