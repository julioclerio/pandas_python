```python
df= pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Valor Venda"].sum()
```

```python
df["Custo"]=df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head()
```

```python
round(df["Custo"].sum(),2)
```

```python
df["Lucro"]=df["Valor Venda"]- df["Custo"]
```

```python
df.head()
```

```python
round(df["Lucro"].sum(),2)
```

```python
df["Tempo envio"] = df["Data Envio"]-df["Data Venda"]
```

```python
df.head()
```

```python
df["Tempo envio"].dtype
```

```python
df["Tempo envio"]=(df["Data Envio"]-df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
df["Tempo envio"].dtype
```

```python
df.groupby("Marca")["Tempo envio"].mean()
```

```python
df.isnull().sum()
```

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"]
```

```python
pd.options.display.float_format="{:20,.2f}".format
```

```python
lucro_ano=df.groupby([df["Data Venda"].dt.year,"Marca"])["Lucro"].sum().reset_index()
```

```python
lucro_ano
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Gráfico de Produtos")
plt.xlabel("Tempo")
plt.xlabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
df_2009=df[df["Data Venda"].dt.year==2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mes 2009")
plt.xlabel("Mes")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca 2009")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro X Classe 2009")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal");
plt.yticks(rotation="horizontal");
```
